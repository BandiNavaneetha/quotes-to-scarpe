```python
import requests
from scrapy import Selector
import pandas as pd
```

```python
res = requests.get('https://quotes.toscrape.com/')
```

```python
sel = Selector(text=res.text)
```

```python
list = []
for i in sel.xpath('//div[@class = "col-md-4 tags-box"]'):
    list.append(i.xpath('.//span/a/text()').extract())

pages = {'Top ten important tags':[]}
for i in range(0,10):
    for r in list:
        pages['Top ten important tags'].append(r[i])
        
df = pd.DataFrame(pages)
df
```

```python
df.to_csv('Top ten important tags.csv',index=False)
```

```python
list = []
for i in sel.xpath('//div[@class = "col-md-4 tags-box"]'):
    list.append(i.xpath('.//span/a/text()').extract())
pages = []
for i in range(0,10):
    for r in list:
        pages.append('https://quotes.toscrape.com/tag/' + str(r[i]))


Quotes = {'Text':[],'Author':[],'Tags':[]}
for j in pages:
    res = requests.get(j)
    sel = Selector(text=res.text)
    for i in sel.xpath('//div[@class = "quote"]'):
        Quotes['Text'].append(i.xpath('.//span/text()').extract_first())
        Quotes['Author'].append(i.xpath('.//span[2]/small/text()').extract_first())
        Quotes['Tags'].append(i.xpath('.//div/a/text()').extract())
```

```python
df = pd.DataFrame(Quotes)
df
```

```python
df.to_csv('tags_scrape.csv',index=False)
```
